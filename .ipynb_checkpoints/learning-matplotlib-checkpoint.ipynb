```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# generate random data using numpy.random
def gen_data(count=17, scale=10, ndim=1):
    arrays = np.array([
        np.sort(np.random.rand(count)) for i in range(0, ndim)
    ]) * scale
    return arrays[0] if ndim is 1 else arrays
```

```python
plt.style.available
```

```python
plt.style.use('ggplot') # see plt.style.available

plt.plot(*gen_data(ndim=2), linewidth=3, label="First Plot") # first plot
plt.plot(*gen_data(ndim=2), 'r--', label="Second Plot") # second plot
plt.plot(*gen_data(ndim=2), color="k", linestyle="dotted", label="Third Plot")

plt.title("The plot title")
plt.xlabel("the x axis")
plt.ylabel("the y axis")

plt.legend() # to show labels of plots
plt.grid(True) # to show coordinates

plt.savefig('./plt.png') # to create an image file

plt.show()
```

```python
y_data = gen_data()
np.random.shuffle(y_data)
plt.style.use('fivethirtyeight')
plt.bar(np.arange(17), y_data)
plt.show()
```
